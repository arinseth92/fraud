````python
# Check merchant records
df['isMerchant'] = df['nameDest'].str.startswith('M')
merchant_missing = df[df['isMerchant']][['oldbalanceDest', 'newbalanceDest']].eq(0).mean()
print("\nPercentage of 0 balances for merchants:\n", merchant_missing)
````

````python
# One-hot encode 'type' column
df = pd.get_dummies(df, columns=['type'], drop_first=True)
````

````python
# Drop columns that are IDs or don't add value for modeling
df.drop(['nameOrig', 'nameDest', 'isMerchant'], axis=1, inplace=True)
````

````python
# Define X (features) and y (target)
X = df.drop(['isFraud'], axis=1)
y = df['isFraud']
````

````python
from sklearn.model_selection import train_test_split

# Split the dataset (stratify = True to maintain fraud ratio)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42)
````

````python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000, class_weight='balanced')
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)
````

````python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
````

````python
from xgboost import XGBClassifier

xgb = XGBClassifier(scale_pos_weight= (len(y_train) - sum(y_train)) / sum(y_train))
xgb.fit(X_train, y_train)

y_pred_xgb = xgb.predict(X_test)

````

````python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Define a function to evaluate each model
def evaluate_model(name, y_true, y_pred, model):
    print(f"=== {name} ===")
    print(confusion_matrix(y_true, y_pred))
    print(classification_report(y_true, y_pred))
    print("ROC AUC Score:", roc_auc_score(y_true, model.predict_proba(X_test)[:,1]))
    print()

# Evaluate
evaluate_model("Logistic Regression", y_test, y_pred_lr, lr)
evaluate_model("Random Forest", y_test, y_pred_rf, rf)
evaluate_model("XGBoost", y_test, y_pred_xgb, xgb)
````

### Q5: What are the key factors that predict fraudulent customer?

Based on feature importance analysis from the Random Forest and XGBoost models, the following features emerged as strong predictors of fraudulent behavior:

1. **Transaction type**  Fraud is mostly associated with `TRANSFER` and `CASH_OUT` operations.
2. **Transaction amount**  Higher transaction amounts are often fraudulent.
3. **oldbalanceOrg** and **newbalanceOrig**  Fraudulent transactions typically empty the senders account.
4. **oldbalanceDest**  The destination account often starts with zero balance, suggesting the use of fresh/mule accounts.

---

### Q6: Do these factors make sense? If yes, How?

Yes, these factors are logical and reflect realistic fraud patterns:

- Fraudsters commonly use **transfer and cash-out methods** to move and withdraw stolen funds.
- They **drain the original account**, leaving the balance at or near zero.
- Destination accounts often have **no prior balance**, indicating new or mule accounts set up for receiving fraudulent funds.

These observations align with real-world fraud strategies such as **account takeovers**, **money laundering**, and **transaction laundering**.
